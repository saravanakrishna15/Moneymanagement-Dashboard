```python
# Generate random data functions

def random_date(start, end):
    return start + timedelta(seconds=random.randint(0, int((end - start).total_seconds())))

def random_description():
    descriptions = ['Tesco Stores 6601', 'Zettle_*donovan?s Bake', 'apple.com/bill', 'Amazon Purchase', 'Uber Trip', 'Netflix Subscription']
    return random.choice(descriptions)

def random_amount():
    return round(random.uniform(-50, -1), 2)

def random_fee():
    return round(random.uniform(0, 10), 2)

def random_balance(prev_balance, amount):
    return round(prev_balance + amount, 2)

# Generate random rows
np.random.seed(0)
random.seed(0)

start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
new_rows = []

for i in range(100):
    last_balance = df.iloc[-1]['Balance']
    started_date = random_date(start_date, end_date)
    completed_date = started_date + timedelta(hours=random.randint(1, 24))
    description = random_description()
    amount = random_amount()
    fee = random_fee()
    currency = 'GBP'
    state = 'COMPLETED'
    balance = random_balance(last_balance, amount)
    
    new_rows.append(['CARD_PAYMENT', 'Current', started_date.strftime('%Y-%m-%d %H:%M:%S'), completed_date.strftime('%Y-%m-%d %H:%M:%S'), description, amount, fee, currency, state, balance])

# Extend the DataFrame
extended_df = pd.concat([df, pd.DataFrame(new_rows, columns=columns)], ignore_index=True)

# Display the first and last 5 rows of the extended DataFrame
print(extended_df.head())
#print(extended_df.tail(30))
```

```python

def calculate_difference(event):
    income = float(income_widget.value)
    recurring_expenses = float(recurring_expenses_widget.value)
    monthly_expenses = float(monthly_expenses_widget.value)
    difference = income - recurring_expenses - monthly_expenses
    difference_widget.value = str(difference)

income_widget = pn.widgets.TextInput(name="Income", value="0")
recurring_expenses_widget = pn.widgets.TextInput(name="Recurring Expenses", value="0")
monthly_expenses_widget = pn.widgets.TextInput(name="Non-Recurring Expenses", value=str(last_month_expenses_tot))
difference_widget = pn.widgets.TextInput(name="Last Month's Savings", value="0")

income_widget.param.watch(calculate_difference, "value")
recurring_expenses_widget.param.watch(calculate_difference, "value")
monthly_expenses_widget.param.watch(calculate_difference, "value")

pn.Row(income_widget, recurring_expenses_widget, monthly_expenses_widget, difference_widget).show()
```

```python
# ## Create monthly expenses trend bar chart 

extended_df['Date'] = pd.to_datetime(extended_df['Date'])            # convert the 'Date' column to a datetime object
extended_df['Month-Year'] = extended_df['Date'].dt.to_period('M')    # extract the month and year from the 'Date' column and create a new column 'Month-Year'
monthly_expenses_trend = extended_df.groupby(['Month-Year', 'Category'])['Amount'].sum().reset_index()

monthly_expenses_trend['Amount']=monthly_expenses_trend['Amount'].astype('str')
monthly_expenses_trend['Amount']=monthly_expenses_trend['Amount'].str.replace('-','')
monthly_expenses_trend['Amount']=monthly_expenses_trend['Amount'].astype('float')
monthly_expenses_trend = monthly_expenses_trend[monthly_expenses_trend["Category"].str.contains("Excluded") == False]

monthly_expenses_trend = monthly_expenses_trend.sort_values(by='Amount', ascending=False)
monthly_expenses_trend['Amount'] = monthly_expenses_trend['Amount'].round().astype(int)
monthly_expenses_trend['Month-Year'] = monthly_expenses_trend['Month-Year'].astype(str)
monthly_expenses_trend = monthly_expenses_trend.rename(columns={'Amount': 'Amount '})

monthly_expenses_trend
```

```python
#Define Panel widget

select_category1 = pn.widgets.Select(name='Select Category', options=[
    'All',
    'Self-Care',
    'Fines',
    'Expenses',
    'Coffee',
    'Groceries',
    'Shopping',
    'Restaurants',
    'Transport',
    'Travel',
    'Entertainment',
    'Gifts',
    'Services',
    #'Excluded'
])

select_category1
```

```python
# define plot function
def plot_expenses(category):
    if category == 'All':
        plot_df = monthly_expenses_trend.groupby('Month-Year').sum()
    else:
        plot_df = monthly_expenses_trend[monthly_expenses_trend['Category'] == category].groupby('Month-Year').sum()
    plot = plot_df.hvplot.bar(x='Month-Year', y='Amount ')
    return plot

# define callback function
@pn.depends(select_category1.param.value)
def update_plot(category):
    plot = plot_expenses(category)
    return plot

# create layout
monthly_expenses_trend_chart = pn.Row(select_category1, update_plot)
monthly_expenses_trend_chart[1].width = 600

monthly_expenses_trend_chart

```
